```python
import pandas as pd
import json

with open('aaa.json', 'r') as f:
    json_data = json.load(f)
print(json.dumps(json_data) )
df = pd.json_normalize(json_data)
df
```

```python
df4 = df["고객질문(요청)"] +df["고객답변"]
df5 = df["상담사답변"] +df["상담사질문(요청)"]
```

```python
df5.drop(0, inplace=True)
df5
```

```python
df5.reset_index()
df111 = pd.DataFrame(df5,columns=['c'])
df111 = df111.reset_index()
df111
```

```python
df4.reset_index()
df112 = pd.DataFrame(df4,columns=['a'])
df112 = df112.reset_index()
df112
```

```python
df112['a'] + df111['c'] 
```

```python
import pandas as pd
import numpy as np
dataset = pd.DataFrame({'q': df112['a'], 'a': df111['c'] })
dataset
dataset['q'].replace('', np.nan, inplace=True)
dataset['a'].replace('', np.nan, inplace=True)
dataset.dropna(subset=['q'], inplace=True)
dataset.dropna(subset=['a'], inplace=True)
dataset
```

```python
dataset.to_csv(r'training.csv', index=False)
```
